# 06 · Cellular Automata (Rules 232 & 30) — learn the mapping


> **TL;DR**  
> This notebook runs a *paired* comparison between **Boolearn** (C, RRR constraints) and PyTorch models trained with **AdamW**, **SGD+Momentum**, and **Muon** (if installed).  
> We measure **generalization** under two modes:
>
> 1) **Black‑box**: flexible MLP + small hyper‑parameter sweep; split the hold‑out into **val**/**test** and pick the best by val.  
> 2) **Principled**: architecture/activations closer to Boolean circuits + L1 to promote **sparsity**; minimal tuning.
>
> **How few training examples** are needed to reach **perfect test accuracy**?
>
> ---
> **Setup expected on your Mac:**  
> * Boolearn repo at `~/Documents/boolearn` (adjust `BOO_LEARN_ROOT` below).  
> * Build Boolearn once: from `src/` run `make all`. (The cells can do this too.)  
> * For Muon (optional): `pip install git+https://github.com/KellerJordan/Muon`.


```

import os, sys, numpy as np, pandas as pd, subprocess
from pathlib import Path

# Path to your local Boolearn checkout
BOO_LEARN_ROOT = os.path.expanduser('~/Documents/boolearn')  # <-- change if needed

# Ensure our shared utils are importable (place this notebook next to experiment_utils.py)
if str(Path('.').resolve()) not in sys.path:
    sys.path.insert(0, str(Path('.').resolve()))

from experiment_utils import *

set_seed(1337)
print("Boolearn root:", BOO_LEARN_ROOT)
```

## Generate Data via `data_cellauto`
- **Rule 232** (majority): size=32, steps=3, expect near-perfect generalization with N≈256.
- **Rule 30** (chaotic): size=16, steps=1, needs deeper architecture.

```

APP = "cellauto"
app_path = os.path.join(BOO_LEARN_ROOT, APP)
os.makedirs(app_path, exist_ok=True)

data_cellauto(BOO_LEARN_ROOT, APP, inbits=3, rule=232, size=32, steps=3, items=10000, out_id="232_3")
with open(os.path.join(app_path, "232_3.wth"), "w") as f:
    f.write("3\n32 32 32 32\n")
layered(BOO_LEARN_ROOT, APP, 0.5, "232_3.wth", "232_3.net")

data_cellauto(BOO_LEARN_ROOT, APP, inbits=3, rule=30, size=16, steps=1, items=10000, out_id="30_1")
with open(os.path.join(app_path, "30_1.wth"), "w") as f:
    f.write("3\n16 32 32 16\n")
layered(BOO_LEARN_ROOT, APP, 0.5, "30_1.wth", "30_1.net")

spec232, X232, Y232 = read_dat(os.path.join(app_path, "232_3.dat"))
spec30,  X30,  Y30  = read_dat(os.path.join(app_path, "30_1.dat"))
print("232:", spec232, X232.shape, Y232.shape)
print("30 :", spec30,  X30.shape,  Y30.shape)
```

## Experiments — Rule 232

```

rng = np.random.RandomState(1337)
N = spec232.n_items
ks = [128, 256]
records232 = []

for k in ks:
    train_idx, val_idx, test_idx = split_train_val_test(N, k, rng)
    order = np.concatenate([train_idx, val_idx, test_idx]).tolist()
    REORDERED = os.path.join(app_path, f"232_3_reordered_k{k}.dat")
    reorder_dat(os.path.join(app_path, "232_3.dat"), REORDERED, order)

    cfg = BoolearnConfig(BOO_LEARN_ROOT, APP, "232_3.net", os.path.basename(REORDERED),
                         items=k+len(val_idx), btfn=3.0, beta=0.2, gam=1e-3,
                         epochs=10, itermax=int(1e6), gapstop=0.1, trials=1,
                         run_id=f"rule232_k{k}")
    res = run_train(cfg)
    bool_test_gen = res["gen_acc"]

    Xtr, ytr = X232[train_idx], Y232[train_idx]
    Xv,  yv  = X232[val_idx],  Y232[val_idx]
    Xte, yte = X232[test_idx], Y232[test_idx]

    grid = [
        dict(hidden=[128], activation="relu", lr=1e-3, weight_decay=1e-4, optimizer_name="adamw"),
        dict(hidden=[128], activation="tanh", lr=3e-3, weight_decay=0.0,  optimizer_name="sgd-momentum"),
        dict(hidden=[256], activation="relu", lr=3e-3, weight_decay=2e-4, optimizer_name="muon"),
    ]
    best_bb = None
    for g in grid:
        tc = TrainConfig(task="binary_vec", epochs=800, batch_size=128, early_stopping_patience=80, l1_lambda=0.0, **g)
        out = train_model(Xtr, ytr, Xv, yv, Xte, yte, tc, verbose=False)
        rec = dict(k=k, mode="blackbox", optimizer=g["optimizer_name"], test_exact=out["test_metrics"]["acc_exact"])
        if best_bb is None or rec["test_exact"] > best_bb["test_exact"]:
            best_bb = rec

    g = dict(hidden=[64,64], activation="hardtanh", lr=3e-3, weight_decay=0.0, optimizer_name="adamw")
    tc = TrainConfig(task="binary_vec", epochs=1200, batch_size=128, early_stopping_patience=120, l1_lambda=1e-6, **g)
    out = train_model(Xtr, ytr, Xv, yv, Xte, yte, tc, verbose=False)
    principled_acc = out["test_metrics"]["acc_exact"]

    records232.append({
        "k": k,
        "boolearn_test_acc": bool_test_gen,
        "bb_best_opt": best_bb["optimizer"],
        "bb_test_exact": best_bb["test_exact"],
        "principled_test_exact": principled_acc
    })

import pandas as pd
pd.DataFrame(records232)
```
